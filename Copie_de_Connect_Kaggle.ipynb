```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#NLTK, Spacy
```

```python

df=pd.read_csv('/content/sample.csv')
```

```python
df
```

```python
textes=df['text']
```

```python
textes
```

```python
import nltk
import re
```

```python
l=textes.str.lower()
```

```python
l
```

```python
import string
l= l.apply(lambda x: x.translate(str.maketrans('', '', string.punctuation)))
```

```python
l
```

```python
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))
l = l.apply(lambda x: ' '.join([word for word in x.split() if word not in stop_words]))
```

```python
l
```

```python
#suppression des mots frequentes et rare
from collections import Counter

# Identifiez les mots fréquents
freq = pd.Series(' '.join(l).split()).value_counts()[:10] # 10 mots les plus fréquents
freq = list(freq.index)
l = l.apply(lambda x: ' '.join([word for word in x.split() if word not in freq]))

# Identifiez les mots rares
rare = pd.Series(' '.join(l).split()).value_counts()[-10:] # 10 mots les moins fréquents
rare = list(rare.index)
l = l.apply(lambda x: ' '.join([word for word in x.split() if word not in rare]))
```

```python
l
```

```python
#Stemming et Lemmatisation
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.tokenize import word_tokenize
nltk.download('punkt')
nltk.download('wordnet')

stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

l= l.apply(lambda x: ' '.join([stemmer.stem(word) for word in word_tokenize(x)]))
l= l.apply(lambda x: ' '.join([lemmatizer.lemmatize(word) for word in word_tokenize(x)]))
```

```python
l
```

```python
# Suppression des émojis et émoticônes
l= l.apply(lambda x: re.sub(r'[\U00010000-\U0010ffff]', '', x))
l= l.apply(lambda x: re.sub(r'[^\x00-\x7F]+', '', x))

# Suppression des URL
l= l.apply(lambda x: re.sub(r'http\S+', '', x))

# Suppression des balises HTML
l= l.apply(lambda x: re.sub(r'<.*?>', '', x))
```

```python
l
```

```python
nouveau_df1 = pd.DataFrame({'tweet_id': df['tweet_id'],'text_cleaned': l,'text':textes})
```
